## Libraries imported

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df= pd.read_csv('Bank_data.csv')
```

```python
df_num = df[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
```

## Visualization using matplotlib

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = 7,5
plt.rcParams['axes.spines.top'], plt.rcParams['axes.spines.right'] = False,False
plt.rcParams['lines.linewidth'] = 2.5
plt.rcParams['xtick.labelsize'] = 'medium'
plt.rcParams['ytick.labelsize'] = 'medium'
```

```python
# Scatter plot
plt.scatter(df['CreditScore'],df['Age'],c='g', s = 10)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph')

# To save the plot:  plt.savefig('plt1.png')
```

```python
# Histograms in subplots
for i in df_num.columns:
    plt.subplots()
    plt.hist(df[i],color='g')
    plt.xlabel(i)
```

```python
# Bar plot, both bar plots will be plotted together
plt.bar(df['NumOfProducts'],df['HasCrCard'])
plt.bar(df['IsActiveMember'],df['Exited'])
```

```python
# Histogram with custom bins
plt.hist(df['Age'],color='g',bins=20)
```

```python
plt.boxplot(df['Balance'],vert = True,patch_artist = True)   # vertical = True for verticalplots, else horizontal plots
plt.show()
```

```python
# Box plots for the numerical columns
for i in df_num.columns:
    plt.subplots()
    plt.boxplot(df[i],vert = True,patch_artist = True)
    plt.xlabel(i)
```

## Seaborn

```python
import seaborn as sns
```

### Numerical value visualization
Types:

Basic: Histograms, Bar plots, Pie charts, Box plots
1. **Corr matrix**
2. **Jointplot** for Bivariate analysis
3. **Distplots** for looking at the distribution of the data in the feature


```python
# Correlation matrix
# More towards blue, means that you have positive correlation
# More towards red indicates negative correlation
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),
            vmax=1, vmin=-1, center=0,
            cmap = sns.diverging_palette(20,220,n=200),
            annot=True)
```

```python
# Bivariate analysis: Let's you analyze two variable in the dataframe
# Here we compare the Balance and Age
sns.jointplot(data=df, x='Balance', y='Age',kind='hex')  # if no kind then it is scatter plot
# if you change kind to reg, then PDF will be placed over the hist
```

```python
 # Pairplot
 sns.pairplot(df, hue = 'Gender')
```

```python
# Distribution plot over the histogram, 
# you can chnage the number of bins
sns.distplot(df['Age'], kde=True, bins=20)
```

### Handling Skewed data using **np.log**
### Other methods: 
###1. Log
###2. Square Root etc.

```python
age = df['Age'].values
age = np.log(age)
```

```python
sns.distplot(age, kde=True, bins=20)

# As you can see the skewness of the data reduced
# The data from a right skewness went to almost a distribution
```

### Categorical data visualization

1.   Countplot, Histograms, Bar plots
2.   Boxplots and Violin plots

Boxplot here is a little different than numerical features

With boxplot, you can only spot outliers when you plot w.r.t some other numerical column



```python
# Simple Count plot
# Given a value, it simply counts the nuber of occurances of the particular category
sns.countplot(df['Exited'])
```

```python
# Bar plot: Supply the x and y values separately
sns.barplot(x='Gender',y='Balance',data=df)
```

```python
# Box plot
sns.boxplot('Gender','Tenure',data=df)
```
